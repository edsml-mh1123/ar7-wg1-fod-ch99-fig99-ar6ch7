# Provenance of IPCC Figures 

This notebook uses the file IPCCFigure.template.trig, which expresses provenance information in PROV-O (TriG format). The Template could be generically applied to any visualisation tool/software and customised to better specify its type (Notebook, Spreasheet, etc) services and dependencies.

ðŸ”— https://www.w3.org/TR/prov-o/ 

For the PROV Template specification used in this example, see:

ðŸ”— https://openprovenance.org/prov-template-2014-06-07/

Below, prov imports and methods are used throughout the notebook to handle and process the provenance data.

```python
from prov.serializers.provjson import ProvJSONSerializer
from prov.serializers.provxml import ProvXMLSerializer
from prov.serializers.provn import ProvNSerializer
from IPython.display import Image
import io
from prov.model import ProvDocument, ProvBundle, ProvException, first, Literal
from prov.dot import prov_to_dot


def provXMLtoVis(doc,format):
     
    
   # doc=ProvXMLSerializer.deserialize(ProvXMLSerializer,doc)
   
    with open(doc, "r", encoding="utf-8") as f:
        prov_doc = ProvDocument.deserialize(f, format="xml")
    dot = prov_to_dot(prov_doc)
    
    if format=="png":
        dot.write_png('PROV.png')
        return 'PROV.png' 
    if format=="svg":
        dot.write_svg('PROV.svg')
        return 'PROV.svg'

def provToVis(doc,informat,format):
     
    
   # doc=ProvXMLSerializer.deserialize(ProvXMLSerializer,doc)
   
    with open(doc, "r", encoding="utf-8") as f:
        prov_doc = ProvDocument.deserialize(f, format=informat)
    
    dot = prov_to_dot(prov_doc)
    
    if format=="png":
        dot.write_png('PROV.png')
        return 'PROV.png' 
    if format=="svg":
        dot.write_svg('PROV.svg')
        return 'PROV.svg'


def provSerializerToN(doc, informat,outputfilename):
     
    with open(doc, "r", encoding="utf-8") as f:
        prov_doc = ProvDocument.deserialize(f, format=informat)
    
    pn=ProvNSerializer(document=prov_doc)
    pn.serialize(open(outputfilename, 'w'))  
```
